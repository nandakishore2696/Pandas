```python
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
delivery.shape
```

```python
delivery.columns
```

```python
delivery.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
new_delivery = delivery[delivery['batsman_runs'] == 4]
new_delivery
```

```python
new_delivery.shape[0] # total number of fours
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
new_delivery.groupby('batsman')['batsman_runs'].size().sort_values(ascending=False).head(10)
```

```python
delivery[delivery['batsman'] == 'V Kohli'].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def runs_scored(name):
    return delivery[delivery['batsman'] == name].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
runs_scored('V Kohli')
```
